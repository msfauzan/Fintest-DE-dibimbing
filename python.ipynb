### Soal 1
#### Hard (Stack & Queue + Pattern)

Seorang pemuda yang penuh semangat bernama **Dibim** berjualan furnitur. Dibim membuka pemesanan tokonya melalui SMS. Format kalimat pemesanan yang diterima yaitu kalimat yang jelas dan konsisten. 1 SMS dikirim oleh 1 orang pelanggan, dan dapat berisi berbagai macam furnitur (>=1 jenis di SMS yang sama) dan jumlah pesanan.
Suatu weekend, Dibim ingin mengikuti bootcamp, alhasil dia ingin anda membantunya untuk membuat pengolah SMS otomatis. Namun karena Dibim adalah anak yang baik, dia juga memberikan kriteria SMS yang dia terima. Jika SMS yang diterima tidak valid, maka Dibim mengabaikan pesan tersebut.

Tugas anda adalah membuat solusi untuk masalah dibim dengan membuat pengolah SMS dan menampilkan furnitur dan jumlah penasan yang dipisahkan oleh spasi. 
- Setiap furnitur terdiri hanya dari 1 kata.
- Jumlah pesanan (angka) bisa terletak setelah atau sebelum nama furnitur, namun tidak keduanya.
- Jika terdapat kata `kecuali` pada suatu kalimat (>=1 kali kemunculan), maka jumlah furnitur yang diambil adalah yang berada di klausa pengecualian setelah klausa utama.
- Apabila formatnya tidak valid atau tidak konsisten, maka ouput yang diharapkan yaitu `abaikan`.
- Jika suatu SMS mengandung kata `kecuali`, maka bisa dipastikan bahwa SMS tersebut valid & konsisten.
- Tuliskan output terurut berdasarkan alfabet nama furnitur.
- Pastikan untuk memberikan 1 baris kosong setelah output utama.
---
Contoh 1:
```
meja 1 kursi 4
```
Output Contoh 1:
```
kursi 4
meja 1

```
---
Contoh 2:
```
meja kursi 4
```
Output Contoh 2:
```
kursi 4
meja 4

```
---
Contoh 3:
```
4 meja 1 kursi
```
Output Contoh 3:
```
kursi 1
meja 4

```
---
Contoh 4:
```
meja 1 2 kursi 3
```
Output Contoh 4:
```
abaikan

```
NB: Diabaikan karena jumlah `kursi` ambigu, antara 3 atau 2.

---
Contoh 5:
```
4 meja kursi 1
```
Output Contoh 5:
```
abaikan

```
NB: Diabaikan karena format pesanan tidak konsisten, jumlah pesanan pertama berada sebelum furnitur, sementara jumlah pesanan kedua berada setelah furnitur.

---
Contoh 6:
```
meja kursi kasur 3 kecuali meja 6
```
Output Contoh 6:
```
kasur 3
kursi 3
meja 6

```
NB: Karena `meja` masuk pengecualian, maka yang dijadikan acuan jumlah yaitu `6`.

---

Notes
- Jika ingin memeriksa apakah suatu kata itu tergolong angka, bisa menggunakan `.isnumeric()`. `True` jika angka, dan `False` jika bukan.
    ```py
    '4'.isnumeric() #True
    'word'.isnumeric() #False
    ```
- Untuk mengurutkan suatu `list`, di Python dapat menggunakan fungsi `sort`. Untuk penggunaan fungsi di dictionary keys, dapat terlebih dahulu menggunakan `items`. I.E 
    ```py
    sorted_dict = dict(sorted(raw_dict.items()))
    # {'a': 1, 'c': 2, 'b': 3} -> {'a': 1, 'b': 3, 'c': 2}
    ```

````python
# Solution
def furniture_solution(sms_text: str):
    def _arr_to_dict(_comps):
        _dict = {}
        for _comp in _comps:
            _key = list(_comp.keys())[0]
            _value = _comp[_key]
            if _key in _dict.keys():
                _dict[_key] += int(_value)
            else:
                _dict[_key] = int(_value)
        return _dict

    def _handle_kecuali(main_comps, kecuali_comps):
        for _key, _value in kecuali_comps.items():
            main_comps[_key] = _value
        return main_comps

    def _number_first(_comps):
        number_stack = []
        furn_stack = []
        result_stack = []
        for _comp in _comps:
            if _comp.isnumeric():
                furn_stack = []
                number_stack.append(int(_comp))
            else:
                furn_stack.append(_comp)
                result_stack.append({_comp: number_stack[-1]})
        if (len(furn_stack) == 0) and (len(number_stack) > 0):
            return "abaikan"
        return _arr_to_dict(result_stack)

    def _furn_first(_comps):
        furn_stack = []
        result_stack = []
        for _comp in _comps:
            if not _comp.isnumeric():
                furn_stack.append(_comp)
            else:
                if len(furn_stack) == 0:
                    return "abaikan"
                result_stack = [
                    *result_stack,
                    *[{_furn: _comp} for _furn in furn_stack],
                ]
                furn_stack = []
        if len(furn_stack) > 0:
            return "abaikan"
        return _arr_to_dict(result_stack)

    def _str_result(result_dict):
        _sorted_res_dict = dict(sorted(result_dict.items()))
        _res_str = ""
        for _key, _value in _sorted_res_dict.items():
            _res_str += f"{_key} {_value}"
            _res_str += "\n"
        return _res_str

    def _main(_text: str):
        _comps = _text.split(" ")
        if "kecuali" in _comps:
            index_of_kecuali = _comps.index("kecuali")
            return _handle_kecuali(
                _main(" ".join(_comps[:index_of_kecuali])),
                _main(" ".join(_comps[index_of_kecuali + 1 :])),
            )
        if _comps[0].isnumeric():
            return _number_first(_comps)
        return _furn_first(_comps)

    _main_result = _main(sms_text)
    _temp_result = "abaikan"
    if _main_result != _temp_result:
        return _str_result(_main_result) 
    return _temp_result + "\n"

_input_examples = [
    "pintu 1 jendela 4",
    "pintu jendela 4",
    "4 pintu jendela",
    "4 pintu jendela 1 bangku",
    "4 pintu 1 jendela",
    "4 pintu jendela 1",  # Invalid
    "pintu 1 2 jendela",  # Invalid
    "pintu 1 2 jendela 3",  # Invalid
    "lampu pintu 1 jendela 3 5",  # Invalid
    "pintu 1 pintu 3",
    "pintu jendela kaca 3 kecuali pintu 6",
    "pintu jendela kaca 3 kecuali pintu 6 kecuali pintu 4",
]
for _input in _input_examples:
    print("~ Test Input      : ")
    print(_input)
    print("~ Expected Output : ")
    print(furniture_solution(_input))
    print("*" * 50)
    print()
````

### Soal 2
#### Hard (Bottom-Up & Dynamic Programming)

Semester ini **Dibim** telah sukses berjualan banyak furnitur berkat bantuan anda. Kini dia ingin coba refreshing dengan melakukan pendakian di Taman Nasional Bukit Barisan Selatan (TNBBS) di Lampung. TNBBS terkenal dengan banyaknya deretan perbukitan yang memiliki ketinggian yang beragam.

Karena Dibim sangat menyukai tantangan, dia penasaran dengan berapa jumlah terbanyak bukit yang ketinggiannya terurut naik. Dibim akan memberikanmu daftar ketinggian tiap bukit, tugasmu adalah untuk memberikan jumlah bukit-bukit yang ketinggiannya terurut naik

- Input berupa daftar ketinggian tiap bukit di TNBBS sejumlah *n* dimana *n>=1*.
- Ketinggian pada suatu bukit *T* independen terhadap masing-masing bukit, dimana *0<=T<10000*
- Ketinggian antar bukit dipisahkan oleh satu karakter spasi, I.E *T1 T2 T3*. 
- Ketinggian antar bukit dinyatakan terurut naik apabila ketinggian *Tj>Ti* dimana *j>i*.
- Pastikan untuk memberikan 1 baris kosong setelah output utama.

---
Contoh 1:
```
1 2 3 4
```
Output Contoh 1:
```
4

```
Jumlah bukit yang terurut naik berjumlah 4, yaitu *(1 2 3 4)*

---
Contoh 2:
```
1 2 5 3 4
```
Output Contoh 2:
```
4

```
Jumlah bukit yang terurut naik terbanyak berjumlah 4, yaitu *(1 2 3 4)*, karena *(1 2 5)*, *(1 2 4)*, & *(1 3 4)* hanya memiliki 3 bukit terurut naik

---
Contoh 3:
```
2 1 4 6 3 5 7 
```
Output Contoh 3:
```
4

```
Jumlah bukit yang terurut naik terbanyak yaitu berjumlah 4, adapun kombinasi yang memungkinkan yaitu
 *(2 4 6 7)*, *(2 4 5 7)*, *(1 4 5 7)*, *(1 4 6 7)*, *(1 3 5 7)*

---
Contoh 3:
```
2 3 3 2 5 7 4 5 6 
```
Output Contoh 3:
```
4

```
Jumlah bukit yang terurut naik terbanyak yaitu berjumlah 5, yaitu *(2 3 4 5 6)*

---

Notes
- Output yang diinginkan yaitu **Jumlah Terbanyak** bukit yang terurut naik, bukan kombinasi dari ketinggian bukit yang terurut naik.
- Untuk mendapatkan full-score, maka perlu diperhatikan terkait memory & time complexity.

````python
# Solution
def bukit_solution(input_tinggi: str):
    arr_tinggi = [int(_t) for _t in input_tinggi.split(" ")]
    arr_max = [1 for _t in input_tinggi.split(" ")]
    for index_i in range(len(arr_tinggi)):
        for index_j in range(index_i):
            if(arr_tinggi[index_i]>arr_tinggi[index_j]):
                arr_max[index_i] = max(arr_max[index_i],arr_max[index_j]+1)
    return str(max(arr_max))+"\n"

_input_examples = [
    "2 3 4 5",
    "2 3 6 5 8 4",
    "3 2 5 7 4 6 7",
    "4 5 5 4 7 9 6 7 8",
    "17 11 4 17 12 8 7 14 18 4 4 13 5",
    "6 20 10 16 15 10 8 3 18 2 16 5 17 5 19 6 10 8",
    "6 17 6 17 20 18 2 15 6 3 17 12 20 13 3 10 14 12 3 9 2 20 7 11",
    "1 7 20 17 18 1 1 1 10 10 9",
    "15 16 7 3 8 18 5 14",
    "12 11 1 17 13 11 3 6",
    "6 4 1 11 4 17 20 12 2 10 4 3 9 5 1 7 20 19 14 7",
    "13 16 17 11 16 2 2 18 17 10 12 1 2 1 20 17 18 15 14 101 10 13 13",
    "10 2 13 9 3 15 8 17 12 17 6 200 3 3 11 1",
    "1 9 3 3 4 11 17 5 12 5 12 20 17 8 6"
]
for _input in _input_examples:
    print("~ Test Input      : ")
    print(_input)
    print("~ Expected Output : ")
    print(bukit_solution(_input))
    print("*" * 50)
    print()
````
